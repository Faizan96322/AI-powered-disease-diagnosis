```python
def get_likely_syms(df,sym_list):
    syms = {}
    i=0
    for sym in sym_list:
        i+=1
        for s in df.columns[:-1]:
            if (s in sym_list):
                continue
            try:
                syms[s] = pd.crosstab(df[sym],df[s])[1][1]
            except:
                print(s)
    syms = sorted(syms.items(),key = lambda x:x[1], reverse=True)
    #print(syms)
    count = int(np.mean([df[df[x]==1].shape[0] for x in sym_list]))
    return [x[0] for x in syms if x[1]>0.6*count]
```

```python
def get_likely_syms_1(df,sym):
    syms = {}
    for s in df.columns[:-1]:
        if s==sym:
            continue
        try:
            syms[s] = pd.crosstab(df[sym],df[s])[1][1]
        except:
            print(s)
    syms = sorted(syms.items(),key = lambda x:x[1], reverse=True)
    #print(syms)
    #count = int(np.mean([df[df[x]==1].shape[0] for x in sym_list]))
    return [x[0] for x in syms if x[1]>0.7*df[df[sym]==1].shape[0]]
```

```python
def get_likely_syms_2(df,sym_list):
    syms = []
    for sym in sym_list:
        sl = {}
        for s in df.columns[:-1]:
            if (s in sym_list):
                continue
            try:
                sl[s] = pd.crosstab(df[sym],df[s])[1][1]
            except:
                print(sym,s)
        sl = sorted(sl.items(),key = lambda x:x[1], reverse=True)
        y = [x[0] for x in sl if x[1]>0] #if x[1]>0.1*df[df[sym]==1].shape[0] 
        syms.append(y)
    #syms = sorted(syms.items(),key = lambda x:x[1], reverse=True)
    #print(syms)
    #count = int(np.mean([df[df[x]==1].shape[0] for x in sym_list]))
    l = syms[0]
    l2=[]
    for s in l:
        flag=1
        for lst in syms[1:]:
            if s in lst:
                flag=1
            else:
                flag=0
                break
        if flag:
            l2.append(s)
    return syms, l2
```
